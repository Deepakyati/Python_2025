```python
#nested dictionaries
students={
    "student1":{"name":"deepak", "age": 30},
    "student2":{"name":"abcd", "age": 35}
}
print(students)
```

```python
##acess nested dict
print(students["student2"]["name"])
print(students["student1"]["age"])
```

```python
#itertaing over dict
for id, info in students.items():
    print(f"{id}:{info}")
    for key,value in info.items():
        print(f"{key}:{value}")
```

```python
#dict comprehension
squares ={x:x**2 for x in range(5)}
print(squares)
```

```python
#conditional dict comprehencion
even_squares ={x:x**2 for x in range(5)}
print(even_squares)
```
