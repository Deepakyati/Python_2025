Temperature conversion

```python
def convert_temp(temp, unit):
    """this func covert temp between celcius and fahrenheit"""

    if unit=="C":
        return temp*9/5 + 32  #celsius to fahrenheit
    elif unit =="F":
        return (temp-32)*5/9
    else:
        return None
    


print(convert_temp(25, "C"))
print(convert_temp(77,"F"))
        
```

Password Strength Checker

```python
def is_strong_pass(password):
    """this func check if pass is strong or not"""
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()' for char in password):
        return False
    return True


##calling function
print(is_strong_pass("weekPWD"))
print(is_strong_pass("Str0ng!@#rt"))
```

Calculate total cost of items in a ashopping cart

```python
def calculate_total(cart):

    total_cost = 0
    for item in cart:
        total_cost+= item["Price"]*item["quantity"]

    return total_cost

cart=[
        {"name": "Apple", "Price":0.5,"quantity":4},
        {"name": "Banana", "Price": 5, "quantity": 12},
        {"name": "Orange", "Price": 25, "quantity": 10}
          
          ]
    
#calling function
total_cost = calculate_total(cart)
print(total_cost)
```

Check if a String is Palindrome

```python
def is_palindrome(st):
    st = st.lower().replace(" ","")
    return st==st[::-1]

print(is_palindrome("abcba"))
print(is_palindrome("Abcba"))
print(is_palindrome("abcbad"))
```

Calculate the factorials of anumber using recursion

```python
def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)
    
print(fact(5))
print(fact(0))
```

Function to read a file and count the frequency of each words

```python
def count_word_freq(file_path):
    word_count = {}
    with open(file_path,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word = word.lower().strip('.,!?;:\"')
                word_count[word]=word_count.get(word,0)+1

    return word_count

filepath='sample.txt' 
word_freq = count_word_freq(filepath)
print(word_freq)
```
