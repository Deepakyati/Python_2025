```python
num=24
if num%2==0:
    print("num is even")
else:
    print("num is odd")
```

```python
#Why function ?
#reusability

def even_odd(num):
    if num%2==0:
        print("num is even")
    else:
        print("num is odd")


#call the function
even_odd(24)
even_odd(19)
```

```python
#function with multiple parameters

def add(a,b):
    #c=a+b
    #return c
    return a+b

add(11,12)
```

```python
#default parameter

def greet(name):
    print(f"Hi {name} how are you ?")

greet("deepak")
```

```python
#variable length arguments
#positional and keywords arguments

#def print_num(num1, num2, num3, num4, num5)

def print_num(*abc):
    for x in abc:
        print(x)

print_num(1,2,3,4,5,6,7,8,9, "deepak", 10)
```

```python
#positional arguments
def print_num(*args):
    for x in args:
        print(x)

print_num(1,2,3,4,5)
```

```python
#kewords arguments
#all the parameter will be in the form of key value pairs

def print_numb(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")

print_numb(name = "deepak", age ="30", country = "India")
```
