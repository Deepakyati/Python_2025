Abstraction

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def drive(self):
        print('this vehicle is used for driving')

    @abstractmethod
    def start_eng(self):
        pass


class Car(Vehicle):
    def start_eng(self):
        print('Car engine started')

def operate_vehicle(vehicle):
    vehicle.start_eng()

car=Car()
operate_vehicle(car)
```
