```python
import pandas as pd

data=[1,2,3,4,5]
series=pd.Series(data)
print(series)
```

```python
#create a series from dict element

data={'a':1,'b':2,'c':3}

seris_dict=pd.Series(data)

print(seris_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
#Dataframe
#create a dataframe from dictioanry of list

data={
    'Name':['Deepak','Deep','Dev'],
    'Age':[25,30,35],
    'City':['A','B','C']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```
